```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for randomness factor at fixed k=4
alpha = [0, 0.2, 0.4, 0.6, 0.8]
novelty_alpha = [4, 7, 5, 6, 5]
feasibility_alpha = [8, 8, 7, 8, 7]
df_alpha = pd.DataFrame({
    'alpha': alpha,
    'novelty': novelty_alpha,
    'feasibility': feasibility_alpha
})

# Data for number of waypoints at fixed alpha=0.2
waypoints = [0, 2, 4, 6, 8]
novelty_k = [5, 8, 7, 8, 8]
feasibility_k = [7, 7, 8, 6, 7]
df_k = pd.DataFrame({
    'waypoints': waypoints,
    'novelty': novelty_k,
    'feasibility': feasibility_k
})

# Plot 1: Impact of randomness factor
plt.figure()
df_alpha.set_index('alpha')[['novelty', 'feasibility']].plot(kind='bar')
plt.xlabel('Randomness Factor (α)')
plt.ylabel('Score')
plt.title('Impact of Randomness Factor on Novelty and Feasibility')
plt.tight_layout()
plt.show()

# Plot 2: Impact of number of waypoints
plt.figure()
df_k.set_index('waypoints')[['novelty', 'feasibility']].plot(kind='bar')
plt.xlabel('Number of Waypoints (k)')
plt.ylabel('Score')
plt.title('Impact of Number of Waypoints on Novelty and Feasibility')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 构造数据
data = {
    'round': list(range(1, 21)),
    'alpha': [1.0,0.7,0.7,0.1,0.8,0.7,0.7,0.7,0.5,0.2,0.8,0.8,0.4,0.0,0.7,0.2,0.2,0.0,0.2,0.4],
    'k':     [2, 8, 9, 8, 8, 2, 3,10, 4, 3, 7, 5, 7, 4, 9, 4, 6, 7, 5, 4],
    'novelty':     [5,5,5,3,6,4,6,3,6,6,6,4,4,6,7,6,5,7,7,9],
    'feasibility': [6,7,7,8,8,7,8,8,8,8,7,7,7,7,8,7,7,8,8,7],
}

df = pd.DataFrame(data)

# Plot 1: Novelty and Feasibility over Rounds
plt.figure(figsize=(12, 4))
plt.plot(df['round'], df['novelty'], marker='o', label='Novelty')
plt.plot(df['round'], df['feasibility'], marker='s', label='Feasibility')
plt.xticks(df['round'])
plt.xlabel('Round')
plt.ylabel('Score')
plt.title('Novelty & Feasibility Scores Over Rounds')
plt.legend()
# plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Parameters α and k over Rounds
fig, ax1 = plt.subplots(figsize=(12, 4))

color = 'tab:blue'
ax1.set_xlabel('Round')
ax1.set_ylabel('α (randomness factor)', color=color)
ax1.plot(df['round'], df['alpha'], color=color, marker='o', label='α')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticks(df['round'])

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('k (waypoints)', color=color)
ax2.plot(df['round'], df['k'], color=color, marker='s', linestyle='--', label='k')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Parameter Values α and k Over Rounds')
# plt.grid(True)
fig.tight_layout()
plt.show()
```
